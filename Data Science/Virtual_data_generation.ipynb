### 분류 모델 학습을 위한 가상 데이터 생성

가상 데이터를 생성하는 이유
* 테스트 이전에 모델의 작동 여부를 확인하기 위해
* 알고리즘의 특성을 이해하거나 성능을 평가하기 위해

##### 1) make_blob을 사용한 가상 데이터 생성
* 정규 분포를 따르는 가상의 데이터 생성
* 여러 개의 클러스터가 존재하는 형태로 데이터 생성 (다중 분류를 위한 클러스터 생성 가능)

```python
from sklearn.datasets import make_blobs

x, y = make_blobs(n_samples=100, centers=3, n_features=2, random_state=1234)

print(x.shape, y.shape)

# n_samples : 생성할 데이터 샘플의 수
# centers : 생성할 클러스터의 수
# n_features : 각 샘플의 feature 갯수
# x : data, y : label
```

```python
# make_blob을 통해 생성한 데이터를 dataframe화
import pandas as pd
import numpy as np

blob_data = pd.DataFrame(dict(x_1 = x[:, 0], x_2 = x[:, 1], label = y))
    # feature 이름을 x_1, x_2라고 가정
blob_data.head()
```

```python
# blob_data의 시각화
import matplotlib.pyplot as plt

plt.scatter(x=blob_data['x_1'], y=blob_data['x_2'], s = 100, c = y)
    # s : datapointer의 크기
    # c : pointer의 색으로, 색이 아닌 label이 주어진 경우, label의 갯수만큼 다른 색 배정
plt.xlabel('X_1')
plt.ylabel('X_2')
plt.show()
```

##### 2) make_moon을 활용한 가상 데이터 생성
* 초승달 모양의 클러스터 2개를 생성 (이진 분류 문제를 위한 데이터셋 제공)
* 비선형 분류 모델을 평가하기 위한 목적으로 사용 : 데이터셋이 비선형적인 결정 경계를 가지기 때문
* 모델의 복잡성 평가 : 비선형 분류 모델이 비선형적 특성을 얼마나 잘 학습하는지 평가하기 위해 사용

```python
# make_moon : 초승달 모양
from sklearn.datasets import make_moons

x, y = make_moons(n_samples=500, noise=0.05, random_state=1234)
    # noise : Gaussian 분포의 표준편차

print(x.shape, y.shape)
```

```python
# make_moon을 통해 생성한 데이터의 dataframe화
moon_data = pd.DataFrame(dict(x_1 = x[:, 0], x_2 = x[:, 1], label = y))

moon_data.head()
```

```python
# moon_data의 시각화
plt.scatter(x=moon_data.x_1, y=moon_data.x_2, s = 50, c = y)
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.show()
```

##### 3) make_circles를 활용한 가상 데이터 생성
* 원 모양의 클러스터 2개를 생성 (이진 분류 문제를 위한 데이터셋 제공)

```python
# circle_data의 시각화
plt.scatter(x=circle_data.x_1, y=circle_data.x_2, s=50, c=y)
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.show()
```

##### 4) make_regression를 활용한 가상 데이터 생성
* regression을 위한 가상 데이터 생성

```python
from sklearn.datasets import make_regression

x, y = make_regression(n_samples=500, n_features=1,
                       n_targets=1, noise=15, random_state=1234)

print(x.shape, y.shape)
```

```python
# make_regression을 통해 생성한 데이터의 dataframe화
regression_data = pd.DataFrame(dict(x = x[:, 0], label = y))

regression_data.head()
```

```python
# regression_data의 시각화
plt.scatter(x=regression_data.x, y=regression_data.label, s=50, c=y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
